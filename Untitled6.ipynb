```python
def predict_risk_for_road(road, city, df_query, model, encoder, query_time=None):
    coords = geocode_road(road, city)

    if coords is None:
        return "Location not found. Please check road and city name."

    lat, lon = coords

    # Use current time if none provided
    if query_time is None:
        query_time = pd.Timestamp.now()

    hour = query_time.hour
    weekday = query_time.weekday()
    month = query_time.month
    is_weekend = int(weekday in [5, 6])
    is_peak_hour = int(hour in [8, 9, 18, 19, 20])  # match your original is_peak_hour definition

    features = pd.DataFrame([{
        "latitude": lat,
        "longitude": lon,
        "hour": hour,
        "is_weekend": is_weekend,
        "is_peak_hour": is_peak_hour,
        "month": month,
        "weekday": weekday
    }])

    pred_encoded = model.predict(features)[0]
    risk_level = encoder.inverse_transform([pred_encoded])[0]

    # Historical context from nearby accident records
    nearby = find_nearby_accidents(df_query, lat, lon)
    accident_count = len(nearby)
    peak_hours = (
        nearby["hour"].value_counts().head(3).to_dict()
        if not nearby.empty else {}
    )

    return {
        "road": road,
        "city": city,
        "predicted_risk_level": risk_level,
        "nearby_accident_count": accident_count,
        "peak_hours_from_history": peak_hours
    }
```
